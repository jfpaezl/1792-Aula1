```python
import pandas as pd
```

```python
datos = pd.read_csv('./data/alquiler.csv', sep=';')
datos.head(10)
```

```python
list(datos['Tipo'].drop_duplicates())
```

```python
recidencial = ['Habitación',
 'Casa',
 'Departamento',
 'Casa en condominio',
 'Casa comercial',
 'Casa de villa',
 ]
recidencial
```

```python
datos.head(10)
```

```python
seleccion = datos['Tipo'].isin(recidencial)
seleccion
```

```python
# filtrar los datos a solo los que son true
datos_residencial = datos[seleccion]
datos_residencial
```

```python
list(datos_residencial['Tipo'].drop_duplicates())
```

```python
#verificar el tamaño de los datos despues de filtrarlos
datos_residencial.shape[0]
```

```python
#verificar el tamaño de los datos antes de filtrarlos
datos.shape[0]
```

```python
# reordenar los indices
datos_residencial.index = range(datos_residencial.shape[0])
datos_residencial
```

# Exportando a la base de datos

```python
# guardar los datos filtrados en un nuevo archivo csv 
datos_residencial.to_csv('./data/alquiler_residencial.csv', sep=';')
```
